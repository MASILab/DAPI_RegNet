```python
mask=np.array(Image.open(mask_path))
```

```python
# Get the instance 7000 in the mask
instance = 7000
mask_instance = np.zeros_like(mask)
mask_instance[mask == instance] = 1

# Check if the instance exists in the mask
if np.any(mask_instance):
    # Calculate the centroid of the instance
    centroid = np.mean(np.argwhere(mask_instance), axis=0)

    # Crop out a square of 100x100 around the centroid
    crop_size = 50  # Half of 100
    crop = mask[int(centroid[0]-crop_size):int(centroid[0]+crop_size), int(centroid[1]-crop_size):int(centroid[1]+crop_size)]

    # Find the boundaries in the cropped mask
    boundaries = segmentation.find_boundaries(crop, mode='inner')

    # Create an RGB image
    rgb_image = np.zeros((crop.shape[0], crop.shape[1], 3), dtype=np.uint8)

    # Set the green channel to the boundaries of the selected instance
    rgb_image[:,:,1] = np.where((boundaries) & (crop == instance), 255, 0)

    # Set the blue channel to the boundaries of the rest of the instances (except 0)
    rgb_image[:,:,2] = np.where((boundaries) & (crop != instance) & (crop != 0), 255, 0)

    # Plot the RGB image
    plt.imshow(rgb_image)
    plt.show()
else:
    print(f"Instance {instance} not found in the mask.")
```
