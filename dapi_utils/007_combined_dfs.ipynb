```python
import pandas as pd
```

```python
df='/fs5/p_masi/rudravg/MxIF_Vxm_Registered/metrics/GCA020TIB_TISSUE01_post_all_V2.csv'
df=pd.read_csv(df)
```

```python
df.head()
```

```python
#Divide this into two dataframes. The first dataframe has the first dataframes contains the frst 29000 rows and the second dataframe contains the rest of the rows. Keep the column names as is
df1 = df.iloc[:29000]
df2 = df.iloc[29000:]
```

```python
df1.head()
```

```python
df2.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np  

# Get the number of columns in df1 (or df2, since they have the same number of columns)
num_columns = len(df1.columns)

# Calculate the number of rows and columns for the subplots
num_rows = int(np.ceil(num_columns / 2))
num_cols = 3 if num_columns > 1 else 1

# Create a figure with a subplot for each column
fig, axs = plt.subplots(num_rows, num_cols, figsize=(20, num_rows*5))

# Flatten the axes array if there's more than one column
if num_cols > 1:
    axs = axs.flatten()

# For each column, create a histogram for df1 and df2 on the same subplot
for i, column in enumerate(df1.columns):
    axs[i].hist(df1[column], bins=50, alpha=0.5, label='df1')
    axs[i].hist(df2[column], bins=50, alpha=0.5, label='df2')
    axs[i].set_title(column)
    axs[i].legend()

# Remove unused subplots if the number of columns is odd
if num_columns % 2 != 0 and num_cols > 1:
    fig.delaxes(axs[-1])

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#For df1 and df2, for every column if the value is greater than 0.5, set it as True and if it is less than 0.5, set it as False
df1_registered = df1 > 0.5
df2_unregistered = df2 > 0.5
```

```python
#Give me the number of True and False for all of the columns as a comparision between df1_registered and df2_unregistered
comparison = pd.concat([df1_registered.sum(), df2_unregistered.sum()], axis=1)
comparison.columns = ['df1', 'df2']
comparison
```

```python
#For df2_unregistered, make a new column called Epi+ and set it as True if nzNorm_Mean_NaKATPase or nzNorm_Mean_PanCK is True
df1_registered['Epi+'] = df1_registered['nzNorm_Mean_NaKATPase'] | df1_registered['nzNorm_Mean_PanCK']
```

```python
df1_registered
```

```python
#For df1_registered, make a new column called Stroma+ and set it as True if nzNorm_Mean_Vimentin or nzNorm_Mean_SMA is True
df1_registered['Stroma+'] = df1_registered['nzNorm_Mean_Vimentin'] | df1_registered['nzNorm_Mean_SMA']
```

```python
df1_registered
```

```python
#For df1_registered, make a column called Exclusion1 and set it as True if Epi+ is True and Stroma+ is True
df1_registered['Exclusion1'] = df1_registered['Epi+'] & df1_registered['Stroma+']
```

```python
df1_registered
```

```python
#Count the number of True for Exclusion1
df1_registered['Exclusion1'].sum()
```

```python
#For df1_registered, make a column called Immune+ and set it as True if nzNorm_Mean_CD20,nzNorm_Mean_CD45,nzNorm_Mean_CD68,nzNorm_Mean_CD11B,nzNorm_Mean_Lysozome,nzNorm_Mean_CD3d,nzNorm_Mean_CD8,nzNorm_Mean_CD4
df1_registered['Immune+'] = df1_registered['nzNorm_Mean_CD20'] | df1_registered['nzNorm_Mean_CD45'] | df1_registered['nzNorm_Mean_CD68'] | df1_registered['nzNorm_Mean_CD11B'] | df1_registered['nzNorm_Mean_Lysozyme'] | df1_registered['nzNorm_Mean_CD3d'] | df1_registered['nzNorm_Mean_CD8'] | df1_registered['nzNorm_Mean_CD4']
```

```python
df1_registered.head()
```

```python
#Removing Immune conflicts for macrohpages
#Make a new column called Exclusion2, set it as True if Exclusion1 is True. Also set it as True if ((CD68 and CD3d is True) or (CD68 and CD20 is True) or (CD68 and CD4 is True) or (CD68 and CD8 is True) or (CD68 and CD11B is True))
df1_registered['Exclusion2'] = df1_registered['Exclusion1'] | ((df1_registered['nzNorm_Mean_CD68'] & df1_registered['nzNorm_Mean_CD3d']) | (df1_registered['nzNorm_Mean_CD68'] & df1_registered['nzNorm_Mean_CD20']) | (df1_registered['nzNorm_Mean_CD68'] & df1_registered['nzNorm_Mean_CD4']) | (df1_registered['nzNorm_Mean_CD68'] & df1_registered['nzNorm_Mean_CD8']) | (df1_registered['nzNorm_Mean_CD68'] & df1_registered['nzNorm_Mean_CD11B']))
```

```python
df1_registered['Exclusion2'].sum()
```

```python
#Remove immune conflics for monocytes
#Make a new column called Exclusion3, set it as True if Exclusion2 is True. Also set it as True if ((CD11B and CD3d is True) or (CD11B and CD20 is True) or (CD11B and CD4 is True) or (CD11B and CD8 is True))
df1_registered['Exclusion3'] = df1_registered['Exclusion2'] | ((df1_registered['nzNorm_Mean_CD11B'] & df1_registered['nzNorm_Mean_CD3d']) | (df1_registered['nzNorm_Mean_CD11B'] & df1_registered['nzNorm_Mean_CD20']) | (df1_registered['nzNorm_Mean_CD11B'] & df1_registered['nzNorm_Mean_CD4']) | (df1_registered['nzNorm_Mean_CD11B'] & df1_registered['nzNorm_Mean_CD8']))
```

```python
df1_registered['Exclusion3'].sum()
```

```python
#Remove immune conflicts for B cells
#Make a new column called Exclusion4, set it as True if Exclusion3 is True. Also set it as True if ((CD20 and CD3d is True) or (CD20 and CD4 is True) or (CD20 and CD8 is True))
df1_registered['Exclusion4'] = df1_registered['Exclusion3'] | ((df1_registered['nzNorm_Mean_CD20'] & df1_registered['nzNorm_Mean_CD3d']) | (df1_registered['nzNorm_Mean_CD20'] & df1_registered['nzNorm_Mean_CD4']) | (df1_registered['nzNorm_Mean_CD20'] & df1_registered['nzNorm_Mean_CD8']))
```

```python
df1_registered['Exclusion4'].sum()
```

```python
#Remove conflicts for helper T and cytotoxic T
#Make a new column called Exclusion5, set it as True if Exclusion4 is True. Also set it as True if ((CD3d False and CD45 False and CD4 is True) or (CD3d False and CD45 False and CD8 True) or (CD4 True and CD8 True))
df1_registered['Exclusion5'] = df1_registered['Exclusion4'] | ((~df1_registered['nzNorm_Mean_CD3d'] & ~df1_registered['nzNorm_Mean_CD45'] & df1_registered['nzNorm_Mean_CD4']) | (~df1_registered['nzNorm_Mean_CD3d'] & ~df1_registered['nzNorm_Mean_CD45'] & df1_registered['nzNorm_Mean_CD8']) | (df1_registered['nzNorm_Mean_CD4'] & df1_registered['nzNorm_Mean_CD8']))
```

```python
df1_registered['Exclusion5'].sum()
```

```python
#Make a new column called Progenitor+ and set it as True if Sox9 or OLFM4 is True
df1_registered['Progenitor+'] = df1_registered['nzNorm_Mean_Sox9'] | df1_registered['nzNorm_Mean_OLFM4']
```

```python
#Make a new column called Exclusion 6 and set it as True if Exclusion 5 is True or (Epi is False and Stroma is False)
df1_registered['Exclusion6'] = df1_registered['Exclusion5'] | (~df1_registered['Epi+'] & ~df1_registered['Stroma+'])
```

```python
df1_registered['Exclusion6'].sum()
```

```python
#Make a new column called Exclusion 7 and set it as True if Exclusion 6 is True or (SMA is True and Immune is True)
df1_registered['Exclusion7'] = df1_registered['Exclusion6'] | (df1_registered['nzNorm_Mean_SMA'] & df1_registered['Immune+'])
```

```python
df1_registered['Exclusion7'].sum()
```

```python
#Make a new column called Exclusion 8 and set it as True if Exclusion 7 is True or (Progenitor is True and Immune is True)
df1_registered['Exclusion8'] = df1_registered['Exclusion7'] | (df1_registered['Progenitor+'] & df1_registered['Immune+'])
```

```python
df1_registered['Exclusion8'].sum()
```

```python
#Make a new column called Exclusion 9 and set it as True if Exclusion 8 is True or (Epi False and Stroma False and Immune False and Progenitor False)
df1_registered['Exclusion9'] = df1_registered['Exclusion8'] | (~df1_registered['Epi+'] & ~df1_registered['Stroma+'] & ~df1_registered['Immune+'] & ~df1_registered['Progenitor+'])
```

```python
df1_registered['Exclusion9'].sum()  
```

```python
#Make a new column called Exclusion 10 and set it as True if Exclusion 9 is True or (Epi True and Immune True)
df1_registered['Exclusion10'] = df1_registered['Exclusion9'] | (df1_registered['Epi+'] & df1_registered['Immune+'])
```

```python
df1_registered['Exclusion10'].sum()
```

```python
#Final annotation for endocrine cells
#Make a new column called Epithelial and set it as True if Epi is True and Exclusion 10 is False
df1_registered['Epithelial'] = df1_registered['Epi+'] & ~df1_registered['Exclusion10']
```

```python
df1_registered['Epithelial'].sum()
```

```python
#Final annotation for stromal cells/fibroblasts
#Make a new column called Stromal and set it as True if Exclusion10 is False, Stroma is True and Immune is False
df1_registered['Stromal'] = ~df1_registered['Exclusion10'] & df1_registered['Stroma+'] & ~df1_registered['Immune+']
```

```python
df1_registered['Stromal'].sum()
```

```python
#Final annotation for fibroblasts
#Make a new column called Fibroblasts and set it as True if Stromal is True and SMA is True and Progenitor False
df1_registered['Fibroblasts'] = df1_registered['Stromal'] & df1_registered['nzNorm_Mean_SMA'] & ~df1_registered['Progenitor+']
```

```python
df1_registered['Fibroblasts'].sum()
```

```python
#Final annotation for stromal nuclei
#Make a new column called Stromal nuclei and set it as True if Stromal is True and SMA is False and Progenitor is False
df1_registered['Stromal nuclei'] = df1_registered['Stromal'] & ~df1_registered['nzNorm_Mean_SMA'] & ~df1_registered['Progenitor+']
```

```python
df1_registered['Stromal nuclei'].sum()  
```

```python
#Final annotation for myleoid nuclei
#Make a new column called Myeloid nuclei and set it as True if Exclusion10 is False and Immune is True and ((Lysozyme True CD68 False and CD11B False and Progenitor False and CD20 False) and (CD3d False CD4 False CD8 False))
df1_registered['Myeloid nuclei'] = ~df1_registered['Exclusion10'] & df1_registered['Immune+'] & ((df1_registered['nzNorm_Mean_Lysozyme'] & ~df1_registered['nzNorm_Mean_CD68'] & ~df1_registered['nzNorm_Mean_CD11B'] & ~df1_registered['Progenitor+'] & ~df1_registered['nzNorm_Mean_CD20']) & (~df1_registered['nzNorm_Mean_CD3d'] & ~df1_registered['nzNorm_Mean_CD4'] & ~df1_registered['nzNorm_Mean_CD8']))
```

```python
df1_registered['Myeloid nuclei'].sum()
```

```python
#Final annotation for helper T nuclei
#Make a new column called Helper T nuclei and set it as True if Exclusion10 is False and Immune is True (CD4 True and Progenitor False)
df1_registered['Helper T nuclei'] = ~df1_registered['Exclusion10'] & df1_registered['Immune+'] & (df1_registered['nzNorm_Mean_CD4'] & ~df1_registered['Progenitor+'])
```

```python
df1_registered['Helper T nuclei'].sum()
```

```python
#Final annotation for cytotoxic T nuclei
#Make a new column called Cytotoxic T nuclei and set it as True if Exclusion10 is False and Immune is True (CD8 True and Progenitor False)
df1_registered['Cytotoxic T nuclei'] = ~df1_registered['Exclusion10'] & df1_registered['Immune+'] & (df1_registered['nzNorm_Mean_CD8'] & ~df1_registered['Progenitor+'])
```

```python
df1_registered['Cytotoxic T nuclei'].sum()
```

```python
#Final annotation for T cell receptors
#Make a new column called T cell receptors and set it as True if Exclusion10 is False and Immune is True (CD3d True and CD4 False and CD8 False)
df1_registered['T cell receptors'] = ~df1_registered['Exclusion10'] & df1_registered['Immune+'] & (df1_registered['nzNorm_Mean_CD3d'] & ~df1_registered['nzNorm_Mean_CD4'] & ~df1_registered['nzNorm_Mean_CD8'])
```

```python
df1_registered['T cell receptors'].sum()
```
